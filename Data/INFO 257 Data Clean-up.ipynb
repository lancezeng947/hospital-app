# Create Visit Table:

```python
visits = pd.read_csv('visit.csv')
small_hospital = pd.read_csv('hospital_table.csv')
```

```python
visits['state'][0] #grab state
# randomly generate a number between 0 and 1, .95 pick instate, otherwise pick completely random 
```

```python
hospital_id = []

for visit_state in visits.state:
    state_hospital_index = small_hospital[small_hospital['state'] == visit_state]['id']

    random_num = np.random.uniform(0,1)
    if random_num <= .85:
        random_hospital = np.random.choice(state_hospital_index)
    else:
        random_hospital = np.random.choice(small_hospital['id'])
    hospital_id.append(random_hospital)
```

```python
visits['hospital_id'] = hospital_id
```

```python
visits.to_csv('visits_temp_table.csv')
```

```python
phys = pd.read_csv('physicians_table.csv')
nurse_table = pd.read_csv('nurse_table.csv')
```

```python
nurse_table
```

```python
nurse_id = []
doctor_id = []

for hos_id in visits.hospital_id:
    phys_index = phys[phys['hospital_id'] == hos_id]['id']
    nurse_index = nurse_table[nurse_table['hospital_id'] == hos_id]['id']

    nurse_id.append(np.random.choice(nurse_index))
    doctor_id.append(np.random.choice(phys_index))
```

```python
visits['doctor_id'] = doctor_id
visits['nurse_id'] = nurse_id
```

```python
visits.to_csv('visits_temp_table.csv')
```

# Drugs & Conditions:

```python
diseases = pd.read_csv('Drugs_Conditions.csv')
```

```python
diseases = diseases[diseases['condition'] != '']
diseases = diseases[diseases['condition'] != 'Birth Control']
diseases.reset_index(drop = True, inplace = True)
diseases['id'] = diseases.index+1
```

```python
diseases = diseases.iloc[0:1500, :].copy()
drugs = diseases.drop_duplicates('drugName').copy()
```

```python
drugs = drugs.loc[:, ['id', 'drugName']]
drugs.columns = ['id', 'name']
```

```python
drugs.reset_index(drop = True, inplace = True)
drugs['id'] = drugs.index+1
```

```python
drugs.to_csv('drugs_table.csv')
```

```python
new_diseases = diseases.loc[:, ['id', 'condition']].copy()
new_diseases = new_diseases.drop_duplicates('condition')
new_diseases.columns = ['id', 'name']
```

```python
new_diseases.reset_index(drop = True, inplace = True)
new_diseases['id'] = new_diseases.index+1
```

```python
new_diseases.to_csv('disease_table.csv')
```

```python
diseases
```

```python
disease_drug = []

for index, row in diseases.iterrows():
    disease_id = new_diseases[new_diseases['name'] == row['condition']]['id'].values[0] #id of the disease
    drug_id = drugs[drugs['name'] == row['drugName']]['id'].values[0]
    
    disease_drug.append((disease_id, drug_id))
```

```python
disease_drug_df = pd.DataFrame(disease_drug)
disease_drug_df.columns = ['disease_id', 'drug_id']
```

```python
disease_drug_df.to_csv('disease_drug_table.csv')
```

```python
merged = diseases.merge(drugs, left_on='drugName', right_on='name')
```

```python
merged.drop(['drugName', 'name'], axis = 1, inplace = True)
merged.columns = ['id', 'condition', 'drug_id']
merged.sort_values('id')
```

```python
merged.to_csv('disease_table.csv')
```

# Insurance Joining Table:

```python
insurance = pd.read_csv('insurance_table.csv')
hospital = pd.read_csv('hospital_table.csv')
```

```python
hospital.shape[0]
```

```python
complete_join = pd.DataFrame()
for index in insurance.id:
    
    insurance_list = []
    
    rand_insurance_num = np.random.randint(10, hospital.shape[0])
    insurance_list = np.array([index]*rand_insurance_num)
    
    temp_df = pd.DataFrame(np.vstack([insurance_list, np.random.choice(hospital['id'], rand_insurance_num)])).T
    
    complete_join = pd.concat([complete_join, temp_df], axis = 0)
```

```python
complete_join.columns = ['insurance_id', 'hospital_id']
```

```python
complete_join.to_csv('insurance_hospital_table.csv')
```
