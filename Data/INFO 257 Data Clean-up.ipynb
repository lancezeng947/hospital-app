# Shuffle Hospitals:


```python
hospitals = pd.read_csv('hospital.csv')
hospitals.head()
```

```python
hospitals = hospitals.sample(frac=1).reset_index(drop=True)
```

```python
small_hospitals = hospitals.iloc[0:1000].copy()
small_hospitals.drop('Provider ID', axis = 1, inplace = True)
```

```python
small_hospitals.to_csv('small_hospitals.csv')
```
